```python
import pandas as pd
import requests
import json
```

```python
state_list = []
start_list = []
end_list = []
gathering_list = []
for x in range(0, 51):
    base_url = 'https://services1.arcgis.com/s6aICfcOJecr58j6/arcgis/rest/services/Stay_at_Home_Orders_and_Advisories/FeatureServer/0/query?where=1%3D1&outFields=*&outSR=4326&f=json'
    data = requests.get(base_url).json()
    state_name = data['features'][x]['attributes']['STATE']
    state_list.append(state_name)
    start_date = data['features'][x]['attributes']['f4']
    start_list.append(start_date)
    end_date = data['features'][x]['attributes']['f5']
    end_list.append(end_date)
    gathering_ban = data['features'][x]['attributes']['f12']
    gathering_list.append(gathering_ban)
```

```python
start_dates = ['n/a' if x==None else x[16:] for x in start_list]
start_dates
```

```python
end_dates = ['n/a' if x==None else x[17:] for x in end_list]
end_dates
```

```python
columns_dict = {
    'state_name': state_list,
    'start_date': start_dates,
    'end_date': end_dates,
    'gathering_bans': gathering_list
}
orders_df = pd.DataFrame(columns_dict)
orders_df.head(30)
```

```python
output_path = 'stay_at_home_orders.csv'
orders_df.to_csv(output_path, index=False)
```
