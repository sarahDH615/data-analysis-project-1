```python
#dependencies
import pandas as pd
import numpy as np
```

```python
#reading in the state abbreviations
path = '../google_trends/statesAbbrev.csv'
statesAbbrev_df = pd.read_csv(path)
statesAbbrev_df.tail(12)
```

```python
#list for state codes
state_codes = [x for x in statesAbbrev_df['Code']]
```

```python
#creating a df dict to hold the state dfs, reading in the dfs
states_df_dict = {}
for x in range(40, 51):
    path = f'../google_trends/google_trends_csvs/state_csvs/US-{state_codes[x]}_cat_data.csv'
    states_df_dict[f'{state_codes[x]}_data_df'] = pd.read_csv(path)
```

```python
col_names_for_max = [x for x in col_names if x not in ['time', 'covid_cases', 'stay_at_home', 'mass_gathering_ban']]
```

```python
ranked_states_40 = states_df_dict[f'{state_codes[40]}_data_df'].iloc[:, 1:22].rank(ascending=False,axis = 1)
```

```python
max_rows = []
max_cols = []
max_indices = []
for x in range(0, len(ranked_states_40)):
    for y in range(0, len(ranked_states_40.columns)):
        if ranked_states_40.iloc[x,y] == 1.0:
            max_rows.append((x))
            max_cols.append((y))
```

```python
print(max_rows)
print(max_cols)
```

```python
max_cols_names = []
for x in max_cols:
    max_cols_names.append(col_names_for_max[x])
```

```python
max_cols_names
```

```python
max_times_names = []
for x in max_rows:
    max_times_names.append(times_list[x])
```

```python
len(max_times_names), len(max_rows), len(max_cols_names)
```

```python
max_pairs = []
for x in range(0, len(max_times_names)):
    max_pairs.append((max_times_names[x], max_cols_names[x]))
max_pairs
```

```python
shared_max_rows = []
shared_max_cols = []
shared_max_indices = []
for x in range(0, len(ranked_states_40)):
    for y in range(0, len(ranked_states_40.columns)):
        if ranked_states_40.iloc[x,y] == 1.5:
            shared_max_rows.append((x))
            shared_max_cols.append((y))
```

```python
len(shared_max_rows)
len(shared_max_cols)
```

```python
shared_max_times_names = []
for x in shared_max_rows:
    shared_max_times_names.append(times_list[x])
```

```python
shared_max_cols_names = []
for x in shared_max_cols:
    shared_max_cols_names.append(col_names_for_max[x])
```

```python
shared_max_pairs = []
for x in range(0, len(shared_max_times_names)):
    shared_max_pairs.append((shared_max_times_names[x], shared_max_cols_names[x]))
shared_max_pairs
```

```python
shared_max_pairs[0][0]
```

```python
shortened_shared_maxes = []
for x in range(0, (len(shared_max_pairs)-1)):
    if shared_max_pairs[x][0] == shared_max_pairs[x+1][0]:
        shortened_shared_maxes.append((shared_max_pairs[x][0], shared_max_pairs[x][1], shared_max_pairs[x+1][1]))
    else:
        continue
shortened_shared_maxes
```
