```python
# import dependencies, and pytrends
from pytrends.request import TrendReq
import pandas as pd
import time

pytrends = TrendReq()
```

```python
terms_2018 = pytrends.top_charts(2018, hl='en-US', tz=300, geo='US')
terms_2018
```

```python
list_2018 = [x for x in terms_2018['title']]
list_2018
```

```python
terms_2019 = pytrends.top_charts(2019, hl='en-US', tz=300, geo='US')
terms_2019
```

```python
list_2019 = [x for x in terms_2019['title']]
list_2019
```

```python
terms_2020 = pytrends.top_charts(2020, hl='en-US', tz=300, geo='US')
terms_2020
```

```python
list_2020 = [x for x in terms_2020['title']]
list_2020
```

```python
top_terms_list = list_2018 + list_2019 + list_2020
top_terms_list
```

```python
searches = ['sports',
 'weather',
 'natural',
 'Disaster', 'disaster',
 'celebrities',
 'celebrity',
 'law', 'police', 'discrimination', 'justice', 'race'
 'action',
 'adventure',
 'film',
 'movie',
 'lottery', 'lotteries',
 'election',
 'video',
 'fantasy',
 'tv', 't.v.',
 'show',
 'smart',
 'mobile',
 'cell',
 'phone',
 'infectious',
 'disease',
 'diseases',
 'virus',
 'health',
 'news',
 'telecom',
 'chat',
 'playstation', 'games', 'consoles']
```

```python
path = 'categories.csv'
categories_df = pd.read_csv(path)
categories_df.head()
```

```python
search_cat_results_names = []
search_cat_results_indices = []
for search in searches:
    print(f'working on {search}')
    for x in range(0, len(categories_df)):
        if search in categories_df.iloc[x, 0].lower():
            search_cat_results_names.append(categories_df.iloc[x, 0])
            search_cat_results_indices.append(categories_df.iloc[x, 1])
```

```python
pot_cat_list = []
for n in top_terms_list:
    try:
        pytrends.build_payload(kw_list=[n])
        related_topic = pytrends.related_topics()
        potential_categories = related_topic[f'{n}']['rising']['topic_type']
        for x in potential_categories:
            pot_cat_list.append(x)
    except KeyError:
        print(n)
        continue
```
