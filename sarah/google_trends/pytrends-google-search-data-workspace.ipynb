```python
from pytrends.request import TrendReq
import pandas as pd

pytrends = TrendReq()
```

```python
path = 'categories.csv'
categs_df = pd.read_csv(path)
categs_df.head()
```

```python
states_path = 'statesAbbrev.csv'
states_abbrev_df = pd.read_csv(states_path)
states_abbrev_df.head()
```

```python
abbrev_list = [f"US-{x}" for x in states_abbrev_df['Code']]
```

```python
cat_list = [x for x in categs_df['category_id']]
```

```python
cat_name_list = [x for x in categs_df['category']]
cat_name_list[0:4]
```

```python
def categ_search(category, category_name, state):
    kw_list = '*'
    try:
        pytrends.build_payload(kw_list, cat=category, timeframe='2019-01-01 2020-12-20', geo=state)
        state_interest_df = pytrends.interest_over_time()
        print(f"{category_name}, {state}")
        print(state_interest_df)
    except IndexError:
        print(f"problem with {category_name}")
```

```python
for x in range(0, len(cat_list)):
    category = cat_list[x]
    category_name = cat_name_list[x]
    for y in range(0, len(abbrev_list)):
        state = abbrev_list[y]
        data = categ_search(category, category_name, state)
```
