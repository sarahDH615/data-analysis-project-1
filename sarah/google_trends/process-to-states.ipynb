```python
#creating a dictionary of all the cat dfs
combo_cats_dict = {
    'arts_entertainment': arts_entertainment_df,
    'autos': autos_vehicles_df,
    'beauty_fitness': beauty_fitness_df,
    'books_lit': books_lit_df,
    'action_adventure': action_adventure_df,
    'campaigns_elections': elections_df,
    'celebs': celebs_df,
    'discrimination': discrim_df,
    'entertainment_media': entertainment_media_df,
    'games_systems_consoles': games_consoles_df,
    'health_news': health_news_df,
    'infectious_diseases': inf_diseases_df,
    'law_enf': law_enf_df,
    'lottos': lotto_df,
    'mobiles': mobiles_df,
    'online_vids': online_vids_df,
    'scifi_fantasy': scifi_fantasy_df,
    'sport_news': sport_news_df,
    'tv_shows': tv_shows_df,
    'voice_vid_chat': vv_chat_df,
    'weather': weather_df
}
```

```python
#list for appending category names to
cats_names_list = []
```

```python
#printing each csv's dict key, appending each to list for columns
for key, value in combo_cats_dict.items():
    print(key)
    #appending cat names to list
    cats_names_list.append(key)
```

```python
#checking that cat list is same length as number of cat dfs (21)
len(cats_names_list)
```

```python
#creating cols for states df dict
cols_for_dict_list = ['time'] + cats_names_list + ['covid_cases'] + ['stay_at_home'] + ['mass_gathering_ban']
```

```python
#creating a dictionary with states as keys, and dfs with cat data for each state as vals
states_info_df_dict = {}
for state in [x for x in states_abbrev_df['state_abbrev']]:
    states_info_df_dict[state] = pd.DataFrame(columns=cols_for_dict_list)
```

```python
#checking length of states df dict - should be same as states list len (51)
len(states_info_df_dict)
```

```python
#appending info to each df and saving it to a csv
for x in range(0, len(states_abbrev_df)):
    target_state = states_abbrev_df.iloc[x, 1]
    for key, value in combo_cats_dict.items():
        states_info_df_dict[target_state]['time'] = value['time']
        states_info_df_dict[target_state][key] = value[target_state]
        states_info_df_dict[target_state]['stay_at_home'] = 'False'
        states_info_df_dict[target_state]['mass_gathering_ban'] = 'False'
    print(states_info_df_dict[target_state])
    output_path = f'google_trends_csvs/state_csvs/{target_state}_cat_data.csv'
    states_info_df_dict[target_state].to_csv(output_path, index=False)
```
