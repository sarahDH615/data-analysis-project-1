```python
state_names = ['Rhode Island', 'Arizona', 'Missouri', 'District of Columbia', 'Idaho', 'Utah', 'New Mexico', 'Colorado', 'Wisconsin', 'Georgia', 'Louisiana', 'Kansas', 'Vermont', 'Minnesota', 'Connecticut', 'Tennessee', 'Indiana', 'Michigan', 'Mississippi', 'Illinois']
previous_orders_start = ['3/30/2020', '3/21/2020', '4/6/2020', '4/1/2020', '3/25/2020', '3/27/2020', '3/23/2020', '3/26/2020', '3/25/2020', '4/3/2020', '3/23/2020', '3/30/2020', '3/24/2020', '3/27/2020', '3/23/2020', '3/31/2020', '3/24/2020', '3/24/2020', '3/31/2020', '3/21/2020']
previous_orders_end = ['4/13/2020', '4/30/2020', '4/24/2020', '4/24/2020', '4/15/2020', '4/13/2020', 'Until further notice', '4/11/2020', '4/24/2020', '4/13/2020', '4/13/2020', '4/19/2020', '4/15/2020', '4/10/2020', '4/22/2020', '4/14/2020', '4/6/2020', '4/13/2020', '4/14/2020', '4/7/2020']
```

```python
prev_orders_dict = {
    'state_name': state_names,
    'previous_orders_start_date': previous_orders_start,
    'previous_orders_end_date': previous_orders_end
}
```

```python
prev_orders_df = pd.DataFrame(prev_orders_dict)
prev_orders_df.head(40)
```

```python
input_path = 'stay_at_home_orders.csv'
stay_home_df = pd.read_csv(input_path)
stay_home_df.head()
```

```python
full_orders_df = pd.merge(stay_home_df, prev_orders_df, on='state_name', how='outer')
full_orders_df.head(52)
```

```python
output_path = 'stay_at_home_orders/full_SAH_orders.csv'
full_orders_df.to_csv(output_path, index=False)
```
